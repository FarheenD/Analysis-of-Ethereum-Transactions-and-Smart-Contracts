# Plot of Number of Transactions vs time

```python
import pandas as pd
from datetime import datetime
from matplotlib import pyplot as plt
import seaborn as sns

list1=[]
with open("transaction_Count.txt") as my_file:
    for line in my_file:
        list1=line.split("],")
        break
        
dict1={}
for l1 in list1:
    l2=l1.replace('[','').replace(']','').split(',')
    date=l2[0].replace('"','').replace(' ','')
    date=datetime.strptime(date, '%Y-%m').strftime("%Y-%m")
    count=int(l2[1])
    dict1[date]=count

sorted_dict_items = sorted(dict1.items(), key=lambda x: x[0])
sorted_dict = {k: v for k, v in sorted_dict_items}

list2=[]
for k,v in sorted_dict.items():
    list2.append([k,v])

df1=pd.DataFrame(list2,columns=["Date","Count"])
ax = df1.plot.bar(x='Date', y='Count', figsize=(18,5))
```

# Plot of Average Transaction value over time

```python
import pandas as pd
from datetime import datetime
from matplotlib import pyplot as plt
import seaborn as sns

list1=[]
with open("transaction_Average.txt") as my_file:
    for line in my_file:
        
        list1=line.split("),")
        break
        
dict1={}
for l1 in list1:
    l2=l1.replace('[','').replace(']','').replace('(','').replace(')','').split(',')
    date=l2[0].replace("'",'').replace(' ','')
    date=datetime.strptime(date, '%Y-%m').strftime("%Y-%m")
    count=float(l2[1])
    dict1[date]=count

sorted_dict_items = sorted(dict1.items(), key=lambda x: x[0])
sorted_dict = {k: v for k, v in sorted_dict_items}

list2=[]
for k,v in sorted_dict.items():
    list2.append([k,v])

df1=pd.DataFrame(list2,columns=["Date","Count"])
ax = df1.plot.bar(x='Date', y='Count', figsize=(18,5))
```
